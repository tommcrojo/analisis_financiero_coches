```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

# Datos
data = {
    'Año': [2021, 2022, 2023],
    'TSLA_Margen_Bruto': [21.02, 25.28, 25.60],
    'VOW3.DE_Margen_Bruto': [18.88, 18.70, 18.93],
    '7203.T_Margen_Bruto': [17.76, 19.03, 16.99],
    'F_Margen_Bruto': [11.32, 15.91, 14.97],
    'GM_Margen_Bruto': [11.16, 14.08, 13.39],
    'TSLA_Margen_Operativo': [6.32, 12.42, 16.76],
    'VOW3.DE_Margen_Operativo': [7.70, 7.92, 7.01],
    '7203.T_Margen_Operativo': [10.89, 9.55, 9.88],
    'F_Margen_Operativo': [6.94, 12.75, 7.85],
    'GM_Margen_Operativo': [7.29, 10.76, 6.58],
    'TSLA_Margen_Beneficio_Neto': [5.41, 4.35, 10.90],
    'VOW3.DE_Margen_Beneficio_Neto': [10.12, 8.30, 4.63],
    '7203.T_Margen_Beneficio_Neto': [5.81, 5.49, 5.87],
    'F_Margen_Beneficio_Neto': [4.37, 4.11, 10.35],
    'GM_Margen_Beneficio_Neto': [9.49, 7.19, 7.36],
    'TSLA_Cobertura_Intereses': [8.87, 33.72, 108.60],
    'VOW3.DE_Cobertura_Intereses': [37.44, 47.76, 21.20],
    '7203.T_Cobertura_Intereses': [131.55, 150.91, 61.94],
    'F_Cobertura_Intereses': [-19.79, -19.00, -39.48],
    'GM_Cobertura_Intereses': [31.22, 39.75, 33.30]
}

# Convertir en un DataFrame
df = pd.DataFrame(data)
```

```python
# Configuración de gráficos
sns.set(style="whitegrid")
```

```python
# Gráfico de Margen Bruto (%)
plt.figure(figsize=(10, 6))
for company in ['TSLA', 'VOW3.DE', '7203.T', 'F', 'GM']:
    plt.plot(df['Año'], df[f'{company}_Margen_Bruto'], marker='o', label=company)
plt.title('Margen Bruto (%)')
plt.xlabel('Año')
plt.ylabel('Margen Bruto (%)')
plt.legend(title='Empresa')
plt.grid(True)
plt.show()
```

```python
# Gráfico de Margen Operativo (%)
plt.figure(figsize=(10, 6))
for company in ['TSLA', 'VOW3.DE', '7203.T', 'F', 'GM']:
    plt.plot(df['Año'], df[f'{company}_Margen_Operativo'], marker='o', label=company)
plt.title('Margen Operativo (%)')
plt.xlabel('Año')
plt.ylabel('Margen Operativo (%)')
plt.legend(title='Empresa')
plt.grid(True)
plt.show()
```

```python
# Gráfico de Margen de Beneficio Neto (%)
plt.figure(figsize=(10, 6))
for company in ['TSLA', 'VOW3.DE', '7203.T', 'F', 'GM']:
    plt.plot(df['Año'], df[f'{company}_Margen_Beneficio_Neto'], marker='o', label=company)
plt.title('Margen de Beneficio Neto (%)')
plt.xlabel('Año')
plt.ylabel('Margen de Beneficio Neto (%)')
plt.legend(title='Empresa')
plt.grid(True)
plt.show()
```

```python
# Gráfico de Cobertura de Intereses
plt.figure(figsize=(10, 6))
for company in ['TSLA', 'VOW3.DE', '7203.T', 'F', 'GM']:
    plt.plot(df['Año'], df[f'{company}_Cobertura_Intereses'], marker='o', label=company)
plt.title('Cobertura de Intereses')
plt.xlabel('Año')
plt.ylabel('Cobertura de Intereses')
plt.legend(title='Empresa')
plt.grid(True)
plt.show()
```
