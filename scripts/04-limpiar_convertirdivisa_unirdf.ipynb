```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
toyota_df=pd.read_csv(r"T:\! Active\Data Analytics\_Projects\04 - Financial Analysis of Car Sector\company_data_csv\7203.T Income Statement (Annual) - Discounting Cash Flows.csv")
tsla_df=pd.read_csv(r"T:\! Active\Data Analytics\_Projects\04 - Financial Analysis of Car Sector\company_data_csv\TSLA Income Statement (Annual) - Discounting Cash Flows.csv")
volkswagen_df=pd.read_csv(r"T:\! Active\Data Analytics\_Projects\04 - Financial Analysis of Car Sector\company_data_csv\VOW3.DE Income Statement (Annual) - Discounting Cash Flows.csv")
ford_df=pd.read_csv(r"T:\! Active\Data Analytics\_Projects\04 - Financial Analysis of Car Sector\company_data_csv\F Income Statement (Annual) - Discounting Cash Flows.csv")
gm_df=pd.read_csv(r"T:\! Active\Data Analytics\_Projects\04 - Financial Analysis of Car Sector\company_data_csv\GM Income Statement (Annual) - Discounting Cash Flows.csv")

display(ford_df)
```

```python
def clean_and_transpose(df, company_name):
    # Eliminamos columna LTM:
    df_cleaned = df.drop(columns=df.columns[1])
    
    # Usamos la primera fila como nombres de columnas
    df_cleaned.columns = df_cleaned.iloc[0]
    df_cleaned = df_cleaned.drop(0).reset_index(drop=True)
    
    # Transponer el Dataframe para tener los años como registros
    df_transposed = df_cleaned.set_index(df_cleaned.columns[0]).T
    df_transposed.index.name = 'Year'
    
    # Extraer solo el año de la fecha y convertirlo a entero
    df_transposed.index = pd.to_datetime(df_transposed.index.str.split('\n').str[0]).year

        # Función para limpiar y convertir valores
    def clean_and_convert(x):
        if isinstance(x, str):
            # Eliminar caracteres no numéricos excepto el punto decimal
            x = ''.join(c for c in x if c.isdigit() or c == '.'  or c == '-')
            try:
                return float(x) * 1_000_000
            except ValueError:
                return np.nan
        elif pd.isna(x):
            return np.nan
        else:
            return float(x) * 1_000_000
    
    # Aplicar la función de limpieza y conversión a todas las columnas numéricas
    for col in df_transposed.columns:
        df_transposed[col] = df_transposed[col].apply(clean_and_convert)
    

    # Añadimos nombre de compañía y lo movemos a la primera posición
    df_transposed.insert(0, "company", company_name)
    
    df_transposed = df_transposed.reset_index(drop=False)
    
    return df_transposed
```

```python
tsla_clean = clean_and_transpose(tsla_df, 'TSLA')
toyota_clean = clean_and_transpose(toyota_df, '7203.T')
volkswagen_clean = clean_and_transpose(volkswagen_df, 'VOW3.DE')
ford_clean = clean_and_transpose(ford_df, 'F')
gm_clean = clean_and_transpose(gm_df, 'GM')

pd.set_option('display.float_format', lambda x: '%.2f' % x)
display(toyota_clean)
```

```python
# Suponiendo que estas son las tasas de conversión:
eur_to_usd = 1.11  # Ejemplo de tasa de EUR a USD
jpy_to_usd = 0.007  # Ejemplo de tasa de JPY a USD

# Para el DataFrame de EUR:
for col in volkswagen_clean.columns[1:]:
    # Verificamos si la columna es numérica
    if pd.api.types.is_numeric_dtype(volkswagen_clean[col]):
        # Convertir los valores de EUR a USD
        volkswagen_clean[col] = volkswagen_clean[col] * eur_to_usd

# Para el DataFrame de JPY:
for col in toyota_clean.columns[1:]:
    if pd.api.types.is_numeric_dtype(toyota_clean[col]):
        toyota_clean[col] = toyota_clean[col] * jpy_to_usd
display(toyota_clean)
```

```python
cardata_a= tsla_clean.merge(toyota_clean, "outer")
cardata_a= cardata_a.merge(volkswagen_clean, "outer")
cardata_a= cardata_a.merge(ford_clean, "outer")
cardata_b= cardata_a.merge(gm_clean, "outer")
display(cardata_b)
```
