```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
toyota_df=pd.read_csv(r"T:\! Active\Data Analytics\_Projects\04 - Financial Analysis of Car Sector\company_data_csv\7203.T Income Statement (Annual) - Discounting Cash Flows.csv")
tsla_df=pd.read_csv(r"T:\! Active\Data Analytics\_Projects\04 - Financial Analysis of Car Sector\company_data_csv\TSLA Income Statement (Annual) - Discounting Cash Flows.csv")
volkswagen_df=pd.read_csv(r"T:\! Active\Data Analytics\_Projects\04 - Financial Analysis of Car Sector\company_data_csv\VOW3.DE Income Statement (Annual) - Discounting Cash Flows.csv")
ford_df=pd.read_csv(r"T:\! Active\Data Analytics\_Projects\04 - Financial Analysis of Car Sector\company_data_csv\F Income Statement (Annual) - Discounting Cash Flows.csv")
gm_df=pd.read_csv(r"T:\! Active\Data Analytics\_Projects\04 - Financial Analysis of Car Sector\company_data_csv\GM Income Statement (Annual) - Discounting Cash Flows.csv")

gm_df.head()
```

```python
def clean_and_transpose(df, company_name):
    df_cleaned = df.drop(columns=df.columns[1])
    
    df_cleaned.columns = [f"{company_name}_{col_name}" for col_name in df_cleaned.iloc[0]]
    df_cleaned = df_cleaned.drop(0).reset_index(drop=True)
    
    df_transposed = df_cleaned.set_index(df_cleaned.columns[0]).T
    df_transposed.index.name = 'Year'
    
    return df_transposed
```

```python
tsla_clean = clean_and_transpose(tsla_df, 'TSLA')
toyota_clean = clean_and_transpose(toyota_df, '7203.T')
volkswagen_clean = clean_and_transpose(volkswagen_df, 'VOW3.DE')
ford_clean = clean_and_transpose(ford_df, 'F')
gm_clean = clean_and_transpose(gm_df, 'GM')

tsla_clean.head()
```

```python
gm_clean.to_csv("gm_data.csv",index=True)
ford_clean.to_csv("ford_data.csv", index=True)
volkswagen_clean.to_csv("volkswagen_data.csv",index=True)
toyota_clean.to_csv("toyota_data.csv",index=True)
tsla_clean.to_csv("tesla_data.csv",index=True)
```
